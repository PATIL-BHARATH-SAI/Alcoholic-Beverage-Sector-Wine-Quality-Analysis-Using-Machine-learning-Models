### Create Quality Categories
 MODIFIED: 3 balanced categories
- ‚â§ 5 ‚Üí Average
- 6-7 ‚Üí Good
- ‚â• 8 ‚Üí Excellent

```python
def quality_category(score):
    if score <= 5:
        return "Average"
    elif score <= 7:
        return "Good"
    else:
        return "Excellent"

df['quality_category'] = df['quality'].apply(quality_category)
```

# Model Building

### Feature & Target Selection

```python
X = df.drop(columns=['quality', 'quality_category','good'], axis=1)
y = df['quality_category']
```

```python
X
```

```python
y
```

### Train‚ÄìTest Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
print("\nTraining set distribution:")
print(y_train.value_counts())
print("\nTest set distribution:")
print(y_test.value_counts())
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

### Handle Class Imbalance with RandomOverSampler
 Using RandomOverSampler instead of SMOTE to avoid k_neighbors issue

```python
from imblearn.over_sampling import RandomOverSampler

print("\nApplying Random Over-Sampling to handle class imbalance...")
ros = RandomOverSampler(random_state=42)
X_train_balanced, y_train_balanced = ros.fit_resample(X_train, y_train)

print("\nAfter Over-Sampling:")
print(pd.Series(y_train_balanced).value_counts())
```

```python
# Scale the balanced data
X_train_balanced_scaled = scaler.fit_transform(X_train_balanced)
```

## Train Models

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
print("\nTraining models with balanced data...")

```

 #### KNN

```python
# Parameter grid for KNN
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

# Initialize KNN
knn = KNeighborsClassifier()

# GridSearchCV
grid_search_knn = GridSearchCV(
    knn,
    param_grid_knn,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)

print("\nStarting GridSearchCV for KNN...")
grid_search_knn.fit(X_train_balanced_scaled, y_train_balanced)

# Results
print("Best parameters for KNN:", grid_search_knn.best_params_)
print("Best accuracy for KNN:", grid_search_knn.best_score_)
```

 #### Logistic Regression

```python
# Parameter grid for Logistic Regression
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l2'],
    'solver': ['lbfgs', 'liblinear'],
    'max_iter': [1000]
}

# Initialize Logistic Regression
lr = LogisticRegression(class_weight='balanced', random_state=42)

# GridSearchCV
grid_search_lr = GridSearchCV(
    lr,
    param_grid_lr,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)

print("\nStarting GridSearchCV for Logistic Regression...")
grid_search_lr.fit(X_train_balanced_scaled, y_train_balanced)

# Results
print("Best parameters for Logistic Regression:", grid_search_lr.best_params_)
print("Best accuracy for Logistic Regression:", grid_search_lr.best_score_)
```

#### SVM

```python
# Parameter grid for SVM
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

# Initialize SVM
svm = SVC(class_weight='balanced', random_state=42)

# GridSearchCV
grid_search_svm = GridSearchCV(
    svm,
    param_grid_svm,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)

print("\nStarting GridSearchCV for SVM...")
grid_search_svm.fit(X_train_balanced_scaled, y_train_balanced)

# Results
print("Best parameters for SVM:", grid_search_svm.best_params_)
print("Best accuracy for SVM:", grid_search_svm.best_score_)
```

#### Decision Tree

```python
# Parameter grid for Decision Tree
param_grid_dt = {
    'max_depth': [None, 5, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

# Initialize Decision Tree
dt = DecisionTreeClassifier(class_weight='balanced', random_state=42)

# GridSearchCV
grid_search_dt = GridSearchCV(
    dt,
    param_grid_dt,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)

print("\nStarting GridSearchCV for Decision Tree...")
grid_search_dt.fit(X_train_balanced_scaled, y_train_balanced)

# Results
print("Best parameters for Decision Tree:", grid_search_dt.best_params_)
print("Best accuracy for Decision Tree:", grid_search_dt.best_score_)
```

#### Random Forest

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# 2. Define a parameter grid for RandomForestClassifier
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10]
}

# 3. Initialize a RandomForestClassifier with class_weight='balanced' and random_state=42
rf_classifier = RandomForestClassifier(class_weight='balanced', random_state=42)

# 4. Instantiate GridSearchCV
grid_search_rf = GridSearchCV(
    estimator=rf_classifier,
    param_grid=param_grid_rf,
    cv=5,
    scoring='accuracy',
    n_jobs=-1, # Use all available cores
    verbose=2
)

# 5. Fit grid_search_rf to X_train_balanced_scaled and y_train_balanced
print("\nStarting GridSearchCV for RandomForestClassifier...")
grid_search_rf.fit(X_train_balanced_scaled, y_train_balanced)

# 6. Print the best parameters found by the grid search
print("\nBest parameters for RandomForestClassifier:", grid_search_rf.best_params_)

# 7. Print the best accuracy score achieved by the best model
print("Best accuracy score for RandomForestClassifier:", grid_search_rf.best_score_)
```

### XGBoost

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

# Define a parameter grid for XGBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder

# Encode target labels to numerical format
label_encoder = LabelEncoder()
y_train_balanced_encoded = label_encoder.fit_transform(y_train_balanced)
y_test_encoded = label_encoder.transform(y_test)

# Using XGBoost for maximum accuracy
xgb_model = XGBClassifier(n_estimators=200, learning_rate=0.1, max_depth=6, random_state=42)
xgb_model.fit(X_train_balanced, y_train_balanced_encoded)
# Using a smaller grid for faster execution in an automated environment
param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 6],
    'learning_rate': [0.05, 0.1],
    'subsample': [0.7, 1.0],
    'colsample_bytree': [0.7, 1.0]
}

# Initialize an XGBClassifier
# Use the encoded balanced target variable
xgb_classifier = XGBClassifier(objective='multi:softmax', eval_metric='mlogloss', use_label_encoder=False, random_state=42)

# Instantiate GridSearchCV
grid_search_xgb = GridSearchCV(
    estimator=xgb_classifier,
    param_grid=param_grid_xgb,
    cv=5,
    scoring='accuracy',
    n_jobs=-1, # Use all available cores
    verbose=2
)

# Fit grid_search_xgb to X_train_balanced_scaled and y_train_balanced_encoded
print("\nStarting GridSearchCV for XGBoostClassifier...")
grid_search_xgb.fit(X_train_balanced_scaled, y_train_balanced_encoded)

# Print the best parameters found by the grid search
print("\nBest parameters for XGBoostClassifier:", grid_search_xgb.best_params_)

# Print the best accuracy score achieved by the best model
print("Best accuracy score for XGBoostClassifier:", grid_search_xgb.best_score_)
```

## Model Comparison

```python
# Get best models
best_knn = grid_search_knn.best_estimator_
best_lr = grid_search_lr.best_estimator_
best_svm = grid_search_svm.best_estimator_
best_dt = grid_search_dt.best_estimator_

# Predictions
knn_pred = best_knn.predict(X_test_scaled)
lr_pred = best_lr.predict(X_test_scaled)
svm_pred = best_svm.predict(X_test_scaled)
dt_pred = best_dt.predict(X_test_scaled)

# Accuracy scores
from sklearn.metrics import accuracy_score

results = {
    "KNN": accuracy_score(y_test, knn_pred),
    "Logistic Regression": accuracy_score(y_test, lr_pred),
    "SVM": accuracy_score(y_test, svm_pred),
    "Decision Tree": accuracy_score(y_test, dt_pred)
}

print("\nTest Set Accuracy Results:")
for model, score in results.items():
    print(f"{model}: {score:.4f}")
```

```python
results = {
    "KNN": grid_search_knn.best_score_,
    "Logistic Regression": grid_search_lr.best_score_,
    "SVM": grid_search_svm.best_score_,
    "Decision Tree": grid_search_dt.best_score_,
    "Random Forest": grid_search_rf.best_score_,
    "XGBoost": grid_search_xgb.best_score_
}

comparison_df = pd.DataFrame.from_dict(
    results, orient='index', columns=['Accuracy']
)

print("Model Comparision")
print(comparison_df)

plt.figure(figsize=(8, 5))
plt.bar(comparison_df.index, comparison_df['Accuracy'], color='skyblue')
plt.title("Model Accuracy Comparison (with Over-Sampling)", fontsize=14)
plt.xticks(rotation=30, ha='right')
plt.ylabel('Accuracy')
plt.tight_layout()
plt.show()
```
